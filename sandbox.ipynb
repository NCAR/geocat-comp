```python
import xarray as xr
import numpy as np
import geocat.comp.meteorology as gc
```

```python
from dask.distributed import LocalCluster, Client
cluster = LocalCluster()
client = Client(cluster)
client
```

```python
ds = xr.open_mfdataset('clm4wa_d02_2011-08-doy225.nc',
    chunks={"south_north":24})
```

```python
u_wind = ds.U.isel(south_north=[0,5], west_east_stag=[0,5], Time=0) # X-wind component (m/s)
v_wind = ds.V.isel(south_north_stag=[0,5], west_east=[0,5], Time=0) # Y-wind component (m/s
q_vapor = ds.QVAPOR.isel(south_north=[0,5], west_east=[0,5], Time=0) # Water vapor mixing ratio (kg/kg)
surface_pressure = ds.PSFC.isel(south_north=[0,5], west_east=[0,5], Time=0) # Surface pressure (Pa)

# We need a constant pressure level system, so going to grab first vertical array for whole system
pressure_lev = ds.P.isel(south_north=0, west_east=0, Time=0) # perturbation pressure (Pa)
```

```python
u_wind = u_wind.rename({"west_east_stag": "west_east"})
v_wind = v_wind.rename({"south_north_stag": "south_north"})
```

```python
delta_pressure = gc.delta_pressure(pressure_lev=pressure_lev, surface_pressure=surface_pressure)
delta_pressure
```

```python
g = 9.80665 # gravitational acceleration (m/s@2)
dpg = delta_pressure / g
dpg.name = "Layer Mass Weighting"
dpg.attrs["units"] = "kg m-2"

dpg
```

```python
uq = u_wind * q_vapor
uq.name = "Zonal Moisture Flux"
uq.attrs["units"] = "m s-1"
uq
```

```python
vq = v_wind * q_vapor
vq.name = "Meridional Moisture Flux"
vq.attrs["units"] = "m s-1"
vq
```

```python
uq_dpg = uq * dpg # mass weighted 'uq'
iuq = uq_dpg.sum(dim="bottom_top")
iuq.name = "Integrated Zonal UQ"
iuq.attrs["units"] = "m s-1"
iuq
```

```python
vq_dpg = vq * dpg # mass weighted 'vq'
ivq = vq_dpg.sum(dim="bottom_top")
ivq.name = "Meridional Zonal VQ"
ivq.attrs["units"] = "m s-1"
ivq
```
